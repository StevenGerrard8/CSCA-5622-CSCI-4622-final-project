```python
#import necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.model_selection import cross_val_score
import seaborn as sns
from skimpy import skim
```

```python
df=pd.read_csv('Android_Malware_Benign.csv')
print("(rows,columns) in data set: "+str(df.shape))
df.head()
```

```python
cnt={}#count the number of non_nan values in each column
for column in df.columns:
    if df[column].count() not in cnt:
        cnt[df[column].count()]=1
    else:
        cnt[df[column].count()]+=1
for i in cnt:
    print("there is "+str(cnt[i])+" columns that have "+str(i)+" non_nan values in the column. ")
```

```python
for column in df.columns:
    cnt=df[column].isna().sum()
    if cnt !=0:
        print(f'column: {column}, number of nans: {cnt}')#count the number of nan value in columns
```

<b>As it is shown above,the data contains 328 columns(327 predictors and 1 response),of which 5 columns each contains 1 nan_value,the others are just 0/1 categorical value.<b>

```python
df=df.dropna()
cnt={}
for column in df.columns:
    if df[column].count() not in cnt:
        cnt[df[column].count()]=1
    else:
        cnt[df[column].count()]+=1
for i in cnt:
    print("there is "+str(cnt[i])+" columns that have "+str(i)+" non_nan values in the column. ")
```

<b>after dropna() method ,there is 4459 rows of data left<b>

```python
df.loc[df['Label']=='Malware','Label']=1 #malware->1
df.loc[df['Label']=='Benign','Label']=0 #benign->0
```

<b>change the response to 0/1 binary value<b>

EDA:

<b> base on the code below , we can show that there is no collinearity (correlation>0.5 ) in the data set. I wish to show the corr matrix but there is too many features so they overlap each other in the heatmap<b>

```python
correlation_matrix = df.corr(numeric_only=True)
high_corr = correlation_matrix > 0.5
corr_matrix = correlation_matrix[high_corr]
np.fill_diagonal(corr_matrix.values, np.nan)
ans=[]
for i in high_corr.columns:
    for j in high_corr.index:
        if high_corr.loc[i,j] is True and (i,j) not in ans and (j,i) not in ans:
          ans.append((i,j))
print(ans)
# plt.figure(figsize=(10, 10))
# sns.heatmap(corr_matrix, annot=True)
# plt.show()
```

Train model:
</b></font>

```python
df=df.astype('int32')
X=df.drop('Label',axis=1)
y=df['Label']#spilt response variable and predictors
dtmodel=DecisionTreeClassifier()
dtscores = cross_val_score(dtmodel, X, y, cv=5)
np.mean(dtscores)
```
